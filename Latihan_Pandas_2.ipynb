```python
import pandas as pd
import numpy as np
```

```python
datas = pd.read_csv('/content/diabetes.csv')
datas.head()
```

```python
datas[datas.Age > 69]
```

```python
datas[~(datas.Pregnancies == 6)].head()
```

```python
datas[(datas.Pregnancies == 3) | (datas.Pregnancies == 0)].head()
```

```python
datas[(datas.Pregnancies > 3) & (datas.Outcome == 0)].head()
```

```python
datas[datas.Pregnancies.isin([2,4])].head()
```

```python
datas_baru = datas[datas.Pregnancies.isin([2, 4, 5, 6]) & (datas.Age > 53)].copy() #JANGAN LUPA COPY(), karena kalo ga pake copy, data yang ada di datas akan ikut
datas_baru.head()
```

```python
datas_baru.sort_values('Pregnancies', ascending=False, inplace=True) #Jangan lupa untuk inplace
datas_baru.head()
```

```python
datas_baru.groupby('Pregnancies').mean()
```

```python
datas_baru.groupby(['Outcome','Pregnancies']).size()
```

```python
datas_baru.groupby(['Outcome','Pregnancies']).mean()
```

```python
datas_baru.pivot_table(index='Outcome', columns='Pregnancies', values='Age', aggfunc='mean')
```
