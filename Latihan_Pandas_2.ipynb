# **MAP, Apply, Transform**

```python
datas_baru['Pregnancies_name'] = datas_baru.Pregnancies.map({2: 'dua', 4: 'empat', 5: 'lima', 6: 'enam'})
datas_baru.head()
```

```python
def encode(x):
  if x <= 60:
    x = 'muda'
  elif x > 60 & x <65:
    x = 'tengah'
  else:
    x = 'tua'
  return x
```

```python
datas_baru['Age_baru'] = datas_baru.Age.apply(encode)
datas_baru.head()
```

```python
datas_baru['Pregnancies_name'] = datas_baru.Pregnancies_name.apply(lambda x: x.upper()) #Kalo mau simple pake lambda
datas_baru.head()
```

```python
datas_baru['Age_scaled'] = datas_baru.Age.transform(lambda x: (x-x.min())/(x.max()-x.min())) #Kalo pake transform, kita langsung masukin seluruh data, ga satu2 kyk .apply()
datas_baru.head()
```

```python
datas_baru = datas_baru.drop(columns=['Pregnancies_name', 'Age_baru', 'Age_scaled'])
datas_baru.head()
```
