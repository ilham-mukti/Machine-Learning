```python
datas = pd.read_csv("/content/gabungan_brand.csv")
datas = pd.get_dummies(datas, columns=['shop_name'])

#filter
shop = ['shop_name_MSI Official Store', 'shop_name_Dell Official Partner']
harga = (3000000, 15000000)
review_count = (40, 450)

datas = datas[datas[shop].any(axis=1) & datas.product_review_count.between(review_count[0], review_count[1]) &
datas.price.between(harga[0], harga[1])]

#
datas.head()
```

```python
class Recsys:
  def __init__(self, datas):
    self.datas = pd.read_csv(datas)
    self.datas = pd.get_dummies(self.datas, columns=['shop_name'])

  def recommend(self, shop=None, harga=None, review_count=None):
    datas = self.datas.copy()
    datas = self.demographic_filter(datas, shop=shop, harga=harga, review_count=review_count)
    datas = self.sorting(datas)
    return datas

  @staticmethod
  def demographic_filter(datas, shop=None, harga=None, review_count=None):
    datas = datas.copy()

    if shop is not None:
      datas = datas[datas[shop].any(axis=1)]
    if harga is not None:
      datas = datas[datas.price.between(harga[0], harga[1])]                         
    if review_count is not None:
      datas = datas[datas.product_review_count.between(review_count[0], review_count[1])]
    return datas

  @staticmethod
  def sorting(datas):
    datas = datas.copy()
    datas = datas.loc[:, 'product_name': 'kelas']
    datas = datas.sort_values('count_sold', ascending=False)
    datas = datas.head(n=6)
    return datas
```

```python
recsys = Recsys('/content/gabungan_brand.csv')
recsys.recommend(shop=['shop_name_MSI Official Store'], harga=(3000000, 15000000),
                 review_count=(10, 950))
```
