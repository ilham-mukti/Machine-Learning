```python
import pandas as pd
```

```python
datas = pd.read_csv("/content/data_skincare_shopee.csv")
datas = pd.get_dummies(datas, columns=['jenis'])

isOfficial = ['isOfficial']
jenis = ['jenis_Facial Wash', 'jenis_Moisturizer']
terjual = (500, 2000)
price = (40000, 100000)

datas = datas[datas[isOfficial].any(axis=1) & datas.terjualCount.between(terjual[0], terjual[1]) &
              datas.price.between(price[0], price[1]) & datas[jenis].any(axis=1)]
datas.head()
```

```python
datas = datas.loc[:, 'nama_product': 'shop_name']
datas = datas.sort_values("terjualCount", ascending=False)
datas.head(10)
```

```python
class Recsys():
  def __init__(self, datas, is_official=None, jenis_skinkare=None, terjual_count=None, price=None):
    self.datas = pd.read_csv(datas)
    self.datas = pd.get_dummies(self.datas, columns=['jenis'])
    self.is_official = is_official
    self.jenis_skinkare = jenis_skinkare
    self.terjual_count = terjual_count
    self.price = price
  
  def recommend(self):
    self.datas = self.demographic_filter()
    result = self.sorting()
    return result

  def demographic_filter(self):
    datas = self.datas.copy()

    if self.is_official is not None:
      datas = datas[datas[self.is_official].any(axis=1)]    
    if self.jenis_skinkare is not None:
      datas = datas[datas[self.jenis_skinkare].any(axis=1)]
    if self.terjual_count is not None:
      datas = datas[datas.terjualCount.between(self.terjual_count[0], self.terjual_count[1])]
    if self.price is not None:
      datas =  datas[datas.price.between(self.price[0], self.price[1])]   
    return datas
    
  
  def sorting(self):
    datas = self.datas.copy()
    datas = datas.loc[:, 'nama_product': 'shop_name']
    datas = datas.sort_values("terjualCount", ascending=False)
    datas = datas.head(10)
    return datas
```

```python
recsys = Recsys(datas='/content/data_skincare_shopee.csv', is_official=['isOfficial'], jenis_skinkare=['jenis_Serum', 'jenis_Moisturizer'],
                terjual_count=(500, 5000), price=(20000, 100000))
recsys.recommend()
```
